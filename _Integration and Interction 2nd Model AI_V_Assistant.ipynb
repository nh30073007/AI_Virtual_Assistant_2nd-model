```python
import speech_recognition as sr
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import pyttsx3
import subprocess

#DATASET
df = pd.read_csv(r'C:\\Users\\nh013\\Desktop\\virtual assistant data\\Bitext_Sample_Customer_Service_Evaluation_Dataset.csv')

# VECTORIZED THE TEXT DATA

vectorizer = TfidfVectorizer()
X = df['utterance']
y = df['intent']
X_vectorized = vectorizer.fit_transform(X)

# TRAIN A MODEL NAIVE BAYS CLASSIFIER
classifier = MultinomialNB()
classifier.fit(X_vectorized, y)

# IINITIALIZE THE TEXT TO SPEECH ENGINE
engine = pyttsx3.init()

# SET   THE   VOICE RATE 
engine.setProperty('rate', 150)  # Adjust the value as desired

# SET    THE    VOICE  VOLUME
engine.setProperty('volume', 0.7)  # Adjust the value between 0 and 1


while True:
   #INITIALIZE THE RECOGNIZER
    r = sr.Recognizer()


    with sr.Microphone() as source:
        print("Listening...")
        audio = r.listen(source)

    try:
        
        text = r.recognize_google(audio)
        print("Recognized Text:", text)

        #VECTORIZED THE RECOGNIZED TEXT
        text_vectorized = vectorizer.transform([text])

        # PREDICT INTENT
        intent = classifier.predict(text_vectorized)
        print("Predicted Intent:", intent[0])


        response = ""  # Default response
        if intent[0] == 'cancel_order':
            response = "Your order cancellation request has been processed."
        elif intent[0] == 'change_order':
            response = "Your order has been successfully modified."
        elif intent[0] == 'check_invoice':
            response ="Here is the invoice for your recent purchase."
        elif intent[0] == 'change_shipping_address':
            response = "your adress has been sucessfully modified."
        elif intent[0] == 'cancel_shipping_address':
            response = "your shipping cancelletion request has been processed."
        elif intent[0] == 'edit_shipping_address':
            response = "your shipping adress successfully modified."
        elif intent[0] == 'change_shipping_address':
            response = "your adress has been sucessfully modified."
        elif intent[0] == 'update_shipping_address':
            response = "your shipping address update sucessfully"
        elif intent[0] == 'recover_shipping_address':
            response = "your shipping address has been recover."
        elif intent[0] == 'check_cancellation_fee':
            response = "five usd."
        elif intent[0] == 'check_payment_methods':
            response = "Here is the payment methods."
        elif intent[0] == 'check_refund_policy':
            response = "Here is the refund policy."
        elif intent[0] == 'contact_customer_service':
            response = "Just wait for a second."
        elif intent[0] == 'delete_account':
            response = "your account has been delete successfully."
        elif intent[0] == 'delivery_options':
            response = "Here is the delivery option."
        elif intent[0] == 'delivery_period':
            response = "wait for 20 minuties"
        elif intent[0] == 'edit_account':
            response = "your account has been modified."
        elif intent[0] == 'get_invoice':
            response = "Here is your invoice."





        engine.say(response)
        engine.runAndWait()

    except sr.UnknownValueError:
        print("Sorry, I didn't understand that.")
    except sr.RequestError:
        print("Sorry, I'm having trouble accessing the speech recognition service.")
```
